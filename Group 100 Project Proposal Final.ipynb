```r
completedt <- rbind(winequalityRed, winequalityWhite) |>
              setNames(c("fixed_acidity",
                         "volatile_acidity",
                         "citric_acid",
                         "residual_sugar",
                         "chlorides",
                         "free_SO2",
                         "total_SO2",
                         "density",
                         "pH",
                         "sulphates",
                         "alcohol",
                         "quality",
                         "type"))
completedt$volatile_acidity <- as.numeric(completedt$volatile_acidity)
completedt$citric_acid <- as.numeric(completedt$citric_acid)
completedt$residual_sugar <- as.numeric(completedt$residual_sugar)
completedt$chlorides <- as.numeric(completedt$chlorides)
completedt$density <- as.numeric(completedt$density)
completedt$sulphates <- as.numeric(completedt$sulphates)
completedt
```

## Methods

We will have plotted our variables in the form of a histogram or boxplot graph. After taking a look at the outcome of both types of graphs, we have concluded that the histograms, in comparison, give us more visually comprehendable results. We plotted histograms relating to each of the variables that contributed in affecting the quality for the wine. Then, we will calculate a correlation coefficient and plot the correlations in order to help visualize the strongest ones. Using the strongest variables, we will use the K-nearest neighbour model to visualize the results. To choose the parameter K, we will use cross validation to compare the classifier’s performance for different values of K, picking the value that correlates to the highest validation accuracy. To visualize the results, we can create a scatter plot with a line graph on it to represent the true values vs the predictions of our model.
